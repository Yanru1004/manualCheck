```python
import docx
from docx.text.paragraph import Paragraph
from docx.table import Table

file_path = r"C:\Users\Devil\Desktop\說明書驗證\TEST_BOOK.docx"

doc = docx.Document(file_path)
```

```python
table_title = ''
for b in doc.element.body:
    #取得區塊類型(p or tbl)
    block_type = b.tag.split('}')[-1]

    if block_type == 'p':
        #取得及記錄標題
        title = Paragraph(b, doc).text
        #跳過空白
        if title.strip() !='':
            table_title = title
        
    elif block_type == 'tbl':
        print(table_title)

        #取得表格物件
        part_list = Table(b,doc)
    
        for n in range(1,len(part_list.rows)):
            if len(part_list.columns) <3:                
                break
                
            a = part_list.cell(n,0).text.strip().replace('\n','')
            b = part_list.cell(n,1).text.strip().replace('\n','')
            c = part_list.cell(n,2).text.strip().replace('\n','')
            d = part_list.cell(n,3).text.strip().replace('\n','')
            print(f"{a};{b};{c};{d}")
            #print(f"{a};{b};{c}")

    else:
        pass
```

```python
for i in range(0,len(doc.tables)):
    
    try:
        print(f'-{i:02d}-')
     
        for n in range(1,len(doc.tables[i].rows)):
            if len(doc.tables[i].columns) <3:
                #print(len(doc.tables[i].columns))
                break
                
            a = doc.tables[i].cell(n,0).text.strip().replace('\n','')
            b = doc.tables[i].cell(n,1).text.strip().replace('\n','')
            c = doc.tables[i].cell(n,2).text.strip().replace('\n','')
            d = doc.tables[i].cell(n,3).text.strip().replace('\n','')
            print(f"{a};{b};{c};{d}")
            #print(f"{a};{b};{c}")
    except IndexError:
        pass
    
    print(' ')
        
```
