```python
import docx

file_path = r"C:\Users\Devil\Desktop\說明書驗證\TEST_BOOK.docx"

doc = docx.Document(file_path)
```

```python
for i in range(0,len(doc.tables)):
    
    try:
        print(f'-{i:02d}-')
     
        for n in range(1,len(doc.tables[i].rows)):
            if len(doc.tables[i].columns) <3:
                #print(len(doc.tables[i].columns))
                break
                
            a = doc.tables[i].cell(n,0).text.strip().replace('\n','')
            b = doc.tables[i].cell(n,1).text.strip().replace('\n','')
            c = doc.tables[i].cell(n,2).text.strip().replace('\n','')
            d = doc.tables[i].cell(n,3).text.strip().replace('\n','')
            print(f"{a};{b};{c};{d}")
            #print(f"{a};{b};{c}")
    except IndexError:
        pass
    
    print(' ')
        
```
